```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM,Input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
df=pd.read_csv("text_emotion_dataset.csv")
text=df["Text"].str.lower()
labels=df["Emotion"]
```

```python
le=LabelEncoder()
y_enc=le.fit_transform(labels)
y_Cat=to_categorical(y_enc)
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(text)
x_seq=tokenizer.texts_to_sequences(text)
```

```python
max_len=max(len(seq)for seq in x_seq)
padded=pad_sequences(x_seq,maxlen=max_len,padding="post")
```

```python
x_train,x_test,y_test,y_train=train_test_split(padded,y_Cat,random_state=42,test_size=0.2)
```

```python
model = Sequential([

    Input(shape=(max_len,)),
    # Input layer with sentence length

    Embedding(
        input_dim=len(tokenizer.word_index)+1,
        output_dim=16
    ),
    # Converts word numbers into 16-dimensional vectors

    LSTM(64),
    # 64 LSTM units
    # Learns sequential patterns in text

    Dense(32, activation="relu"),
    # Hidden layer for deeper learning

    Dense(len(le.classes_), activation="softmax")
    # Output layer
    # Softmax gives probability for each emotion
])
```

```python
model.compile(
    optimizer="adam",                  # Adaptive optimizer
    loss="categorical_crossentropy",   # Multi-class classification loss
    metrics=["accuracy"]               # Track accuracy
)
```

```python
model.fit(
    x_train, y_train,
    epochs=30,             # Train 30 times
    batch_size=4,          # 4 samples per batch
    validation_split=0.2,  # 20% of training used for validation
    verbose=1              # Show training progress
)
```

```python
# =====================================
# Text Emotion Detection using LSTM
# =====================================

import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

# Load dataset
df = pd.read_csv("text_emotion_dataset.csv")

texts = df["Text"].str.lower()
labels = df["Emotion"]

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(labels)
y_cat = to_categorical(y_encoded)

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

X_seq = tokenizer.texts_to_sequences(texts)

# Padding
max_len = max(len(seq) for seq in X_seq)
X_padded = pad_sequences(X_seq, maxlen=max_len, padding='post')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_padded, y_cat, test_size=0.2, random_state=42
)

# Build LSTM model
model = Sequential([
    Input(shape=(max_len,)),
    Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=16),
    LSTM(64),
    Dense(32, activation="relu"),
    Dense(len(le.classes_), activation="softmax")
])

# Compile model
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Train model
model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=4,
    validation_split=0.2,
    verbose=1
)

print("\nText Emotion LSTM Model Trained Successfully")

# Evaluate
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
def predict_emotion(text):
    text=text.lower()
    seq=tokenizer.texts_to_sequences([text])
    padded=pad_sequences(seq,maxlen=max_len,padding="post")
    prediction=model.predict(padded,verbose=0)
    pred_index=np.argmax(prediction)
    return le.inverse_transform([pred_index])[0]

sample_text = "I feel very sad today"

print("Text:", sample_text)
print("Predicted Emotion:", predict_emotion(sample_text))
```
