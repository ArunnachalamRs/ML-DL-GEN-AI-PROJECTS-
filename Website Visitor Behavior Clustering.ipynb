```python
import pandas as pd

data = [
    [5, 3, 10, 80, 30],
    [12, 6, 25, 60, 55],
    [20, 10, 40, 45, 70],
    [3, 2, 6, 90, 20],
    [30, 15, 60, 35, 85],
    [25, 12, 50, 40, 80],
    [4, 3, 8, 85, 25],
    [15, 8, 30, 55, 60],
    [28, 14, 55, 38, 82],
    [10, 5, 20, 65, 50],
    [35, 18, 70, 30, 90],
    [8, 4, 15, 75, 40],
]

columns = [
    "SessionDuration",
    "PagesVisited",
    "Clicks",
    "BounceRate",
    "ScrollDepth"
]

df = pd.DataFrame(data, columns=columns)
df.to_csv("website_visitors.csv", index=False)

print("Dataset created successfully: website_visitors.csv")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_absolute_error

from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("website_visitors.csv")
df.head(5)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Train K-Means model
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)
kmeans.fit(X_scaled)
```

```python
df["Cluster"] = kmeans.labels_

print("\nClustered Data:")
print(df)

# View cluster centers
print("\nCluster Centers (scaled values):")
print(kmeans.cluster_centers_)
```

```python
sample_visitor = [[
    22,  # SessionDuration
    11,  # PagesVisited
    45,  # Clicks
    42,  # BounceRate
    75   # ScrollDepth
]]

sample_scaled = scaler.transform(sample_visitor)
cluster = kmeans.predict(sample_scaled)

print("\nNew Visitor belongs to Cluster:", cluster[0])
```
