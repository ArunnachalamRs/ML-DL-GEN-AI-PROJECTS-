```python
import pandas as pd

data = [
    [35, 30, 10, 1012, 0, "Sunny"],
    [33, 35, 12, 1010, 0, "Sunny"],
    [28, 60, 15, 1008, 2, "Cloudy"],
    [25, 70, 18, 1005, 5, "Rainy"],
    [22, 80, 20, 1002, 10, "Rainy"],
    [30, 55, 14, 1009, 1, "Cloudy"],
    [36, 28, 9, 1013, 0, "Sunny"],
    [27, 65, 16, 1007, 3, "Cloudy"],
    [24, 75, 19, 1004, 8, "Rainy"],
    [34, 32, 11, 1011, 0, "Sunny"],
    [29, 58, 15, 1008, 2, "Cloudy"],
    [23, 82, 22, 1001, 12, "Rainy"],
]

columns = [
    "Temperature",
    "Humidity",
    "WindSpeed",
    "Pressure",
    "Rainfall",
    "Condition"
]

df = pd.DataFrame(data, columns=columns)
df.to_csv("weather_data.csv", index=False)

print("Dataset created successfully: weather_data.csv")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
df=pd.read_csv("weather_data.csv")
df.head(6)
```

```python
le=LabelEncoder()
df["Condtion"]=le.fit_transform(df["Condition"])
```

```python
X=df.drop("Condition",axis=1)
y=df["Condtion"]
```

```python
y_cat=to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_cat, test_size=0.25, random_state=42
)
```

```python
scaler= StandardScaler()
X_train_sca=scaler.fit_transform(X_train)
X_test_sca=scaler.fit(X_test)
```

```python
model = Sequential([
    Dense(16, activation="relu", input_shape=(X_train_sca.shape[1],)),
    Dense(12, activation="relu"),
    Dense(3, activation="softmax")  # 3 weather classes
])
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X_train_sca,y_train,validation_split=0.2,epochs=20,batch_size=4,verbose=1)
```

```python

# Custom prediction
sample_weather = [[
    26,  # Temperature
    68,  # Humidity
    17,  # WindSpeed
    1006,# Pressure
    4    # Rainfall
]]

sample_scaled = scaler.transform(sample_weather)
prediction = model.predict(sample_scaled)
predicted_class = np.argmax(prediction)

print(
    "\nPredicted Weather Condition:",
    le.inverse_transform([predicted_class])[0]
)
```
