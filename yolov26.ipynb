```python
from ultralytics import YOLO

# 1. Load the YOLO26 model
model = YOLO("yolo26n.pt")

# 2. Train the model
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,
    imgsz=640,
    device=0,      # Use 0 for GPU, or 'cpu'
    optimizer='MuSGD', # YOLO26 specific optimizer for better convergence
    plots=True
)
```

```python
# Load your custom trained model
my_model = YOLO("/content/runs/detect/train/weights/best.pt")

# Run detection on an image
results = my_model.predict(source="/content/License-Plate-Object-Detection-6/test/images/1998_California_license_plate_4BJW994_jpg.rf.688d38335c6f7fc559437587a51ee6cd.jpg", save=True, conf=0.25)

# Process results
for result in results:
    boxes = result.boxes
    for box in boxes:
        print(f"Detected Plate at: {box.xyxy}") # Bounding box coordinates
```
