```python
import pandas as pd

data = {
    "Question": [
        "hi",
        "hello",
        "how are you",
        "what is your name",
        "bye",
        "thank you",
        "good morning",
        "good night"
    ],
    "Answer": [
        "hello",
        "hi there",
        "i am fine",
        "i am a chatbot",
        "goodbye",
        "you are welcome",
        "good morning",
        "sweet dreams"
    ]
}

df = pd.DataFrame(data)
df.to_csv("chatbot_dataset.csv", index=False)

print("Dataset created successfully: chatbot_dataset.csv")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("chatbot_dataset.csv")
```

```python
questions=df["Question"].str.lower()
answers=df["Answer"].str.lower()
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(questions)
```

```python
X_seq=tokenizer.texts_to_sequences(questions)
```

```python
max_len= max(len(seq) for seq in X_seq)
x=pad_sequences(X_seq,maxlen=max_len,padding="post")
```

```python
lable_encoder=LabelEncoder()
y=lable_encoder.fit_transform(answers)
```

```python
num_cls=len(set(y))
```

```python
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index)+1,
              output_dim=16,
              input_length=max_len),
    SimpleRNN(32),
    Dense(num_cls, activation="softmax")
])
```

```python
model.compile(
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
    optimizer="adam"
)
```

```python
model.fit(x,y,epochs=100,verbose=1)
```

```python
def chatbot(user_imput):
    user_imput=user_imput.lower()
    seq=tokenizer.texts_to_sequences([user_imput])
    padded=pad_sequences(seq,maxlen=max_len,padding="post")

    prediction=model.predict(padded,verbose=0)
    pred_index=np.argmax(prediction)

    response=lable_encoder.inverse_transform([pred_index])
    return response[0]

while True:
    user_imput=input("you: ")
    if user_imput.lower()=="exit":
        break
    print("Bot:",chatbot(user_imput))
```
